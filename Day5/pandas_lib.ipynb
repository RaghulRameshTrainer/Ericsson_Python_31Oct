```python
def get_language(city):
    if city in ['Chennai','Podicherry']:
        return "Tamizh"
    elif city in ["Mumbai","Delhi"]:
        return "Hindi"
    elif city=="Calcutta":
        return "Bengali"
    elif city=='Cochin':
        return "Malayalam"
    elif city=='Hyderabad':
        return 'Telugu'
    else:
        return "HINDI"
    
```

```python
df['Language']=df['Place'].apply(get_language)
```
